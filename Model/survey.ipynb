Now, we focus on our target: **Salary_Last_Year** and remove features inconsequential to this.

*The above is based on the fact that, subjective reasons and statements do not influence salary: for example Reason or the question: 1. What is your biggest reason for learning to code? does not affect salary earned due to the fact that this question is based on psychologial inclination towards preferred job.*

```python
df1 = df.copy().drop(columns=['Timestamp','Reason','Reasons-SDE','interested Careers'])
```

changing salary to numerical

```python
import numpy as np

for i in range(len(df1['Salary_Last_Year'])):
    value = df1['Salary_Last_Year'][i]
    if value in ["I don't want to answer", "I don’t know", np.nan]:
        df1.loc[i, 'Salary_Last_Year'] = np.random.choice([0, 1])
    elif value in ['$1,000 to $2,999', '$15,000 to $19,999', '$5,000 to $6,999', '$7,000 to $9,999', 'Under $1,000', '$3,000 to $4,999', '$10,000 to $14,999', '$20,000 to $24,999']:
        df1.loc[i, 'Salary_Last_Year'] = 0
    else:
        df1.loc[i, 'Salary_Last_Year'] = 1
```

changing disabilities benefit to numerical

```python
df1['Disability_Benefits-Government'] = df1['Disability_Benefits-Government'].map({'Yes': 1, 'No': 0})

median_value = df1['Disability_Benefits-Government'].median()

df1['Disability_Benefits-Government'].fillna(median_value, inplace=True)
```

changing debts and combining them for total debt

```python
debt_columns = ['Debt Car_Loans', 'Debt Credit_Cards', 'Debt Medical_Loans', 
                'Debt Mortgages', 'Debt Payday_Loans', 'Debt Student_Loans', 'Debt Other']

for j in debt_columns:
    for i in range(len(df1[j])):
        value = df1[j][i]
        if value == "<$10k":
            df1.loc[i, j] = 5000
        elif value == "$10k- $25k":
            df1.loc[i, j] = 15000
        elif value == "$26k- $50k":
            df1.loc[i, j] = 40000
        elif value == "$51k- $100k":
            df1.loc[i, j] = 80000
        elif value == "$101k- $200k":
            df1.loc[i, j] = 160000 
        elif value == "$201k- $500k":
            df1.loc[i, j] = 400000
        elif value == ">$500k":
            df1.loc[i, j] = 600000
        else:
            df1.loc[i, j] = 0
```

```python
df1[debt_columns] = df1[debt_columns].astype(float)

df1['Total Debt'] = df1[debt_columns].sum(axis=1)

df1 = df1.drop(columns=debt_columns,axis=1)
```

```python
df1['High-Speed_Internet-Home'] = df1['High-Speed_Internet-Home'].map({'Yes': 1, 'No': 0})

median_value = df1['High-Speed_Internet-Home'].median()

df1['High-Speed_Internet-Home'].fillna(median_value, inplace=True)
```
